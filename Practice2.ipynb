```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

Original_image=cv2.imread("/content/img.bmp",1)
image=cv2.imread("/content/img.bmp",0)

plt.imshow(Original_image,cmap="gray")
plt.title('Original Image')
plt.axis("off")
plt.show()
```

```python
# Equation of Negation=255-image

neg_img=255-image

plt.imshow(neg_img,cmap="gray")
plt.title('Negative Image')
plt.axis("off")
plt.show()
```

```python
# Equation of Log Transformation=c*log(1+r)

log_img=np.log1p(image)

plt.imshow(log_img,cmap="gray")
plt.title('Log Transformation Image')
plt.axis("off")
plt.show()
```

```python
# Equation of Power Log Transformation=c(r^gamma)

for i in range(4):
  gamma=0.5*i
  plt.subplot(2,2,i+1)
  plt.imshow(np.power(image,gamma),cmap="gray")
  plt.title('Power Log Transformation Image')
  plt.axis("off")
plt.show()
```

```python
horizontal_flip=cv2.flip(image,1)
plt.imshow(horizontal_flip,cmap='gray')
plt.title('Horizontal Flipped Image')
plt.axis("off")
plt.show()
```

```python
vertical_flip=cv2.flip(image,0)
plt.imshow(vertical_flip,cmap='gray')
plt.title('Vertical Flipped Image')
plt.axis("off")
plt.show()
```
