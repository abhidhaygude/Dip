```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

#Avrage Filter

original_image=cv2.imread("/content/cameraman.bmp",1)
original=cv2.imread("/content/cameraman.bmp",0)

padded = np.pad(original, ((1,1),(1,1)), mode='constant').astype(float)

output = np.zeros_like(original, dtype=float)

rows,cols=original.shape

for i in range(rows):
  for j in range(cols):
    s=0;
    for k in range(3):
            for l in range(3):
              s=s+padded[i+k][j+l]
    s=s/9
    output[i][j]=s

output = np.uint8(output)

plt.subplot(1,2,1)
plt.imshow(original_image,cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(output,cmap='gray')
plt.title("Avrage Filter Output")
plt.axis("off")

plt.show()


```

```python
#Weighted Filter

mask=[[1,2,1],[2,4,2],[1,2,1]]

padded= np.pad(original,((1,1),(1,1)),mode='constant').astype(float)

output=np.zeros_like(original,dtype='float')

rows,cols = original.shape

for i in range(rows):
  for j in range(cols):

    s=0;

    for k in range(3):
            for l in range(3):
                s=s+padded[i+k][j+l]*mask[k][l]

    output[i][j]=s/16

output=np.uint8(output)

plt.subplot(1,2,1)
plt.imshow(original_image,cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(output,cmap='gray')
plt.title("Weighted Filter Output")
plt.axis("off")

plt.show()
```

```python
# Median Filter

output = original.copy()

padded= np.pad(original,((1,1),(1,1)),mode='constant').astype(float)

rows, cols = original.shape

for i in range(rows):
  for j in range(cols):
    arr=[]

    for k in range(3):
      for l in range(3):
        arr.append(padded[i+k][j+l])

    arr.sort()
    output[i][j]=arr[4]

plt.subplot(1,2,1)
plt.imshow(original, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(output, cmap='gray')
plt.title("Median Filter Output")
plt.axis("off")

plt.show()
```
