```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

dark=cv2.imread("/content/Dark.bmp",1)

plt.subplot(1,2,1)
plt.imshow(dark,cmap='gray')
plt.title("Dark Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(dark.ravel(), bins=256, range=(0,256))
plt.title("Histogram")

plt.show()
```

```python
uneql_hist=cv2.imread("/content/Low.bmp",1)
hist=cv2.imread("/content/Low.bmp",0)

eq_hist=cv2.equalizeHist(hist)

unequal_hist=cv2.calcHist([uneql_hist],[0],None,[256],[0,256])
equal_hist=cv2.calcHist([eq_hist],[0],None,[256],[0,256])

plt.subplot(2,2,1)
plt.imshow(uneql_hist,cmap='gray')
plt.title("Unequalized Image")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(eq_hist,cmap='gray')
plt.title("Equalized Image")
plt.axis('off')

plt.subplot(2,2,3)
plt.plot(unequal_hist)
plt.title("Unequalized Histogram")

plt.subplot(2,2,4)
plt.plot(equal_hist)
plt.title("Equalized Histogram")
```
